# Dealing with the data

### Show the dataframe shape

```python
### Importing dataframe and displaying shape

import pandas as pd
df1 = pd.read_csv("./files_for_lab/csv_files/marketing_customer_analysis.csv")
display(df1.head())
df1.shape
```

### Standardize header names

```python
### Way number 1 with list comprehension

#df1.columns = [column.lower() for column in df1.columns]
#df1.head()

### Way number 2 with lambda functions

df1.columns = list(map(lambda x: x.lower(), df1.columns))
df1.head()
df1 = df1.iloc[:,1:]
df1.head()
```

### Which columns are numerical?

```python
print(df1.dtypes)
df1.select_dtypes("number")
```

### Which columns are categorical?

```python
df1.select_dtypes("object")
```

### Check and deal with NaN values

```python
print(df1.isna().sum())
all(df1[df1["state"].isna() == True].index == df1[df1["response"].isna() == True].index)
all(df1[df1["months since last claim"].isna() == True].index == df1[df1["number of open complaints"].isna() == True].index)
all(df1[df1["vehicle class"].isna() == True].index == df1[df1["vehicle size"].isna() == True].index)

### Since the NAs in the above mentioned columns seem to be more or less in the same rows, I will just get rid of these NA values 
```

```python

### Here I am deleting the NAs in these columns 

df2 = df1.copy()

nan_columns = ["state", "months since last claim", "vehicle class"]

for column in nan_columns:
    df2 = df2[df2[column].isna() == False]


print(df2.isna().sum())
print(df2.shape)
df2.head()
```

```python
### Here I will see what to do with the column that has so many missing values 

df2["vehicle type"].unique()

### It seems to have only nan and "A", so I will just replace the nan with "A". However, this column wont give me any useful information in the future
```

```python
df2['vehicle type'] = df2['vehicle type'].fillna("A")
print(df2['vehicle type'].unique())
print(df2.shape)
```

### Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.

```python
df2["effective to date"] = pd.to_datetime(df2["effective to date"], errors = "coerce")
print(df2.dtypes)
display(df2.head())
print(df2.shape)
```

```python
df3 = df2.copy()

df3["month"] = pd.DataFrame(map(lambda x: x.month, df3["effective to date"]))
display(df3.head())
print(df3.dtypes)
print(df3["month"].value_counts())
print(df3.shape)
```

```python
def clean_month(x):
    if x == 1.0:
        return "January"
    elif x == 2.0:
        return  "February"


df3["month"] = list(map(clean_month, df3["month"]))
print(df3["month"].value_counts())
print(df3.shape)

### There seem to be no other months other than January or February 
```
